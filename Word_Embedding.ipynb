```python
from keras.preprocessing.text import one_hot
```

```python
sent =['I ate dinner',
        'We had a three-course meal',
        'Brad came to dinner with us',
        'He loves fish tacos',
        'In the end, we all felt like we ate too much',
        'We all agreed; it was a magnificent evening',
        'I hope that I will be able to travel to Mexico',
        'it would be lovely to enjoy a week soaking up the culture',
        'Of all the places to travel, Mexico is at the top of my list',
        'Would you like to travel with me',
        'There is so much to understand',
        'I love learning',
        'Sentences come in many shapes and sizes',
        'Nothing beats a complete sentence']
```

```python
voc_size=10000
```

```python
onehot=[one_hot(words,voc_size) for words in sent]
```

```python
onehot
```

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
```

```python
sent_length=14
padded_onehot=pad_sequences(onehot,maxlen=sent_length,padding='pre')
```

```python
padded_onehot
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(padded_onehot))
```
